7. Рассмотрим случайный вектор с формой 100х2, представляющей координаты, и найдем расстояния от точки к точке

````python
points = np.random.rand(100, 2)

def distance(point1, point2):
    return np.sqrt(np.sum((point1 - point2) ** 2))

distances = np.zeros((100, 100))
for i in range(100):
    for j in range(100):
        distances[i, j] = distance(points[i], points[j])

print(distances[7, 3])
````

````python
import pandas as pd
````

````python
DATA_PATH = "russian-housing-market.csv"
df = pd.read_csv(DATA_PATH)
df.head()
````

1. Посчитайте количество записей в датасете с отсутствующей информацией о жилой площади жилого помещения

````python
print(df.shape[0])
missing = df['life_sq'].isnull().sum()
print(missing)
````

2. Предложите способ заменить отсутствующие значения из пункта 1 и реализуйте его. Способ нужно аргументировать(сказать почему это достаточно хорошее приближение)

````python
# жилая площадь часто зависит от района. то есть мы можем найти среднее по
# району нежилое пространство и вычислить
# его. А потом вычитать из общей площади
# нежилое пространство для пропущенных
# значений в life_sq

df['non_living_space'] = df['full_sq'] - df['life_sq']
df['avg_non_living_space'] = df.groupby('sub_area')['non_living_space'].transform('mean')

df.loc[df['life_sq'].isna(), 'life_sq'] = (
    df.loc[df['life_sq'].isna(), 'full_sq'] -
    df.loc[df['life_sq'].isna(), 'avg_non_living_space']
)
missing = df['life_sq'].isnull().sum()
print(missing)
````

### 3. Визуализация

Используя библиотеки matplotlib/sns/plotly/pandas и набор данных представленный ранее визуализируйте

````python
import matplotlib.pyplot as plt
````

1. Как менялась средняя цена квартиры в Нагатинском Затоне(Nagatinskij Zaton) с течением времени

````python
df_nz = df[df['sub_area'] == 'Nagatinskij Zaton'].copy()

df_nz['timestamp'] = pd.to_datetime(df_nz['timestamp'])
df_nz_grouped = (
    df_nz.groupby(pd.Grouper(key='timestamp', freq='Y'))['price_doc']
    .mean()
    .reset_index()
)

plt.figure(figsize=(12, 6))
plt.plot(df_nz_grouped['timestamp'], df_nz_grouped['price_doc'])
plt.xlabel('Дата')
plt.ylabel('Средняя цена')
plt.title('Изменение средней цены квартир в Нагатинском Затоне')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````

2. Постройте гистограмму распределения цен на квартиры с общей площадью более 50 кв.метров

````python
filtered_df = df[df['full_sq'] > 50].copy()

plt.figure(figsize=(12, 6))
filtered_df['price_doc'].hist(bins=30)
plt.xlabel('Цена квартиры, десятки млн.руб.')
plt.ylabel('Количество квартир')
plt.title('Распределение цен на квартиры площадью более 50 кв.м.')
plt.grid()
plt.show()
````

3. Визуализируйте количество проданных квартир в разных районах

````python
district_counts = df.groupby('sub_area')['id'].count().reset_index()

district_counts = district_counts.sort_values('id', ascending=False)

plt.figure(figsize=(25, 6))
plt.bar(district_counts['sub_area'], district_counts['id'])
plt.xticks(rotation=90)
plt.xlabel('Район')
plt.ylabel('Количество проданных квартир')
plt.title('Количество проданных квартир в разных районах')
plt.grid()
plt.show()
````

7. Постройте график изменения медианной цены на квартиру менее 50 кв. метров за период с 2011 до 2014 года*

\* Задание на доп. балл

````python
df_small = df[
    (df['life_sq'] < 50) &
    (df['timestamp'] >= '2011-01-01') &
    (df['timestamp'] <= '2014-12-31')
].copy()

df_small['timestamp'] = pd.to_datetime(df_small['timestamp'])
df_small_grouped = (
    df_small.groupby(pd.Grouper(key='timestamp', freq='M'))['price_doc']
    .median()
    .reset_index()
)

plt.figure(figsize=(12, 6))
plt.plot(df_small_grouped['timestamp'], df_small_grouped['price_doc'])
plt.xlabel('Дата')
plt.ylabel('Медианная цена')
plt.title('Изменение медианной цены на квартиры площадью менее 50 кв. м (2011-2014)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
````
